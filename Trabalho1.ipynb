```python
import pandas as pd #manipulação de dados em formato de dataframe
import numpy as np #biblioteca para operações matemáticas multidimensionais
import seaborn as sns #biblioteca de visualização de informações estatísticas
import matplotlib.pyplot as plt #biblioteca de visualização de dados
from scipy import stats
from google.colab import drive
```

```python
drive.mount('/content/drive')

# Caminho do diretório escolhido
path = "/content/drive/MyDrive/Colab Notebooks"
```

```python
# Leitura do arquivo
df_profissao = pd.read_csv(path + '/age_dataset.csv', sep=',')

# tirando valores nulos da area de estudo de interesse

df_profissao = df_profissao.dropna(subset= ['Gender', 'Occupation', 'Age of death'] )


```

```python
#tabela com as 4 profissões que contém mais mulheres.
df_profissao= df_profissao[['Gender', 'Occupation', 'Age of death']]

mulheres = df_profissao[df_profissao['Gender' ] == 'Female']


mulheres['Occupation'].value_counts().head(4)


```

```python
 # pegando dados apenas de mulheres artistas
 mulheres_artistas = mulheres[mulheres['Occupation'] == 'Artist']
```

```python
# Definir o número de classes e gerar o histograma
num_classes = 6
dados = mulheres_artistas['Age of death']

frequencia, limites_classes = np.histogram(dados, bins=num_classes)

# Arredondar os limites das classes
limites_classes = np.round(limites_classes)

# Frequência Relativa (usando SciPy)
frequencia_relativa = stats.relfreq(dados, numbins=num_classes)

# Ponto Médio de cada classe (arredondado)
pontos_medios = np.round((limites_classes[:-1] + limites_classes[1:]) / 2)

# Frequência Acumulada
frequencia_acumulada = np.cumsum(frequencia)

# Criando uma tabela organizada
tabela_final = pd.DataFrame({
    'Limite Inferior': limites_classes[:-1],
    'Limite Superior': limites_classes[1:],
    'Frequência': frequencia,
    'Frequência Relativa': np.round(frequencia_relativa.frequency, 4),  # Arredonda para 4 casas decimais
    'Frequência Acumulada': frequencia_acumulada,
    'Ponto Médio': pontos_medios
})

# Exibir a tabela
print(tabela_final)

```

```python
moda = stats.mode(mulheres_artistas['Age of death'])

print(f"A moda é: {moda}")

media = mulheres_artistas['Age of death'].mean()

print(f"A média é: {media}")

mediana = mulheres_artistas['Age of death'].median()

print(f"A mediana é: {mediana}")
```

```python
# Histograma
plt.figure(figsize=(15,10))
sns.histplot(data=mulheres_artistas, x='Age of death',log_scale=False, bins=20, color='darkorchid')
plt.xlabel('Idade', fontsize=10)
plt.ylabel('Frequência', fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='Age of death', data= mulheres_artistas, color='lightblue')
plt.title('Boxplot da Idade')
plt.ylabel('Idade')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
# Calcular a média e a moda da idade
media_idade = mulheres_artistas['Age of death'].mean()
moda_idade = mulheres_artistas['Age of death'].mode()[0]  # Pega a primeira moda
mediana_idade = mulheres_artistas['Age of death'].median()
# Criar um boxplot da idade
plt.figure(figsize=(8, 6))
sns.boxplot(y='Age of death', data= mulheres_artistas, color='lightblue')

# Adicionar linha tracejada da média
plt.axhline(media_idade, color='red', linestyle='--', label=f'Média: {media_idade:.2f}')
# Adicionar linha tracejada da moda
plt.axhline(moda_idade, color='green', linestyle='--', label=f'Moda: {moda_idade}')
#
plt.axhline(mediana_idade, color='blue', linestyle='--', label=f'Mediana: {mediana_idade}')

plt.title('Boxplot da Idade')
plt.ylabel('Idade da Morte')
plt.legend()

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
